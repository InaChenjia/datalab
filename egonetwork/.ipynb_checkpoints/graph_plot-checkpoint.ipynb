```python
try:
    import cPickle as pickle
except ImportError:
    import pickle

f = open('call_graph_dump.txt', 'rb')
call_graph = pickle.load(f)
f.close()
```

```python
import networkx as nx
print nx.info(call_graph)
```

```python
samples = map(lambda x: x.strip('\r\n'), open('call_samples.txt', 'r').readlines())
print samples
```

```python
import networkx as nx
center = samples[1]
G = call_graph
#eg1 = nx.ego_graph(G, center, radius = 1)
#eg2 = nx.ego_graph(G, center, radius = 2)
eg3 = nx.ego_graph(G, center, radius = 3)
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(8,8))
pos=nx.spring_layout(eg3)
nx.draw_networkx_nodes(eg3,pos,node_size=12)
nx.draw_networkx_edges(eg3,pos,edge_color='r') 
#nx.draw_networkx_labels(eg3,pos,fontsize=5)
plt.show()
```
